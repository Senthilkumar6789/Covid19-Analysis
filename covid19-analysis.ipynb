```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
data = pd.read_csv("/kaggle/input/covid19/country_wise_latest.csv")
data.head()
```

```python
data_sorted_by_Country = data.sort_values(by='Country/Region', ascending=False)
data_sorted_by_Deaths = data.sort_values(by='Deaths', ascending=False)
data_sorted_by_Recovered = data.sort_values(by='Recovered', ascending=False)
```

```python
top_ten_Country_data = data_sorted_by_Country.head(10)
top_ten_Deaths_data = data_sorted_by_Deaths.head(10)
top_ten_Recovered_data = data_sorted_by_Recovered.head(10)
```

```python
sns.barplot(x="Country/Region", y="Deaths", data=top_ten_Deaths_data)
sns.lineplot(x="Country/Region", y="Deaths", data=top_ten_Deaths_data)
plt.title("Countries with maximum deaths")
plt.xlabel("Country/Region")
plt.ylabel("Deaths")
plt.xticks(rotation=45)
plt.show()
```

```python
# Comparison between Deaths and Recovered for each country
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.barplot(x="Country/Region", y="Deaths", data=top_ten_Deaths_data, color='blue')
plt.title("Deaths")
plt.xlabel("Country/Region")
plt.ylabel("Deaths")
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.barplot(x="Country/Region", y="Recovered", data=top_ten_Recovered_data, color='green')
plt.title("Recovered")
plt.xlabel("Country/Region")
plt.ylabel("Recovered")
plt.xticks(rotation=45)

plt.tight_layout()

plt.show()
```
